```python
import os, glob, numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import keras.backend.tensorflow_backend as K

import tensorflow as tf
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
session = tf.Session(config=config)

X_train_s1, X_train_s2, X_train_s3, X_train_s4, X_train_s5, X_train_s6, X_train_s7, X_train_s8, X_train_s9, X_train_s10, X_train_s11, X_train_s12, X_test_s1, X_test_s2, X_test_s3, X_test_s4, X_test_s5, X_test_s6, X_test_s7, X_test_s8, X_test_s9, X_test_s10, X_test_s11, X_test_s12, y_train, y_test = np.load('C:/Datasets/train/multi_image_data.npy')
print(X_train_s1.shape)
print(X_train_s1.shape[0])
```

```python
categories = ["B7","B9","B10","B13","B14","B15","B16", "B17" ,"C"]
nb_classes = len(categories)

#일반화
X_train_s1 = X_train_s1.astype(float) / 255
X_train_s2 = X_train_s2.astype(float) / 255
X_train_s3 = X_train_s3.astype(float) / 255
X_train_s4 = X_train_s4.astype(float) / 255
X_train_s5 = X_train_s5.astype(float) / 255
X_train_s6 = X_train_s6.astype(float) / 255
X_train_s7 = X_train_s7.astype(float) / 255
X_train_s8 = X_train_s8.astype(float) / 255
X_train_s9 = X_train_s9.astype(float) / 255
X_train_s10 = X_train_s10.astype(float) / 255
X_train_s11 = X_train_s11.astype(float) / 255
X_train_s12 = X_train_s12.astype(float) / 255
X_test_s1 = X_test_s1.astype(float) / 255
X_test_s2 = X_test_s2.astype(float) / 255
X_test_s3 = X_test_s3.astype(float) / 255
X_test_s4 = X_test_s4.astype(float) / 255
X_test_s5 = X_test_s5.astype(float) / 255
X_test_s6 = X_test_s6.astype(float) / 255
X_test_s7 = X_test_s7.astype(float) / 255
X_test_s8 = X_test_s8.astype(float) / 255
X_test_s9 = X_test_s9.astype(float) / 255
X_test_s10 = X_test_s10.astype(float) / 255
X_test_s11 = X_test_s11.astype(float) / 255
X_test_s12 = X_test_s12.astype(float) / 255
```

```python
from keras.models import Model
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, Input, concatenate

sensor=[]
x=[]
for i in range(12):
    sensor.insert(i,Input(shape=(64,64,3)))

for i in range(12):
    x.insert(i,Conv2D(filters=32, kernel_size=3, activation=tf.nn.relu, padding='SAME')(sensor[i]))
    x[i]=MaxPooling2D(padding ='SAME')(x[i])
    x[i]=Conv2D(filters=64, kernel_size=3, activation=tf.nn.relu, padding='SAME')(x[i])
    x[i]=MaxPooling2D(padding ='SAME')(x[i])
    x[i]=Conv2D(filters=128, kernel_size=3, activation=tf.nn.relu, padding='SAME')(x[i])
    x[i]=MaxPooling2D(padding ='SAME')(x[i])
    x[i]=Flatten()(x[i])
    x[i]=Dense(256, activation = tf.nn.relu)(x[i])
    x[i]=Dropout(0.5)(x[i])
    x[i]=Dense(64, activation = tf.nn.relu)(x[i])
    x[i]=Dropout(0.5)(x[i])
    x[i]=Dense(1)(x[i])
    x[i]= Model(inputs=sensor[i], outputs=x[i])

result = concatenate([x[0].output,x[1].output, x[2].output, x[3].output, x[4].output, x[5].output, x[6].output, x[7].output, x[8].output, x[9].output, x[10].output, x[11].output])

output = Dense(12, activation="relu")(result)
output = Dense(9, activation="softmax")(output)


model = Model(inputs=[x[0].input, x[1].input, x[2].input, x[3].input, x[4].input, x[5].input, x[6].input, x[7].input, x[8].input, x[9].input, x[10].input, x[11].input],outputs=output)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_dir = './model'

if not os.path.exists(model_dir):
    os.mkdir(model_dir)
    
model_path = model_dir + '/multi_img_classification.model1'
checkpoint = ModelCheckpoint(filepath=model_path , monitor='val_loss', verbose=1, save_best_only=True)
early_stopping = EarlyStopping(monitor='val_loss', patience=6)
```

```python
model.summary()
from keras.utils import plot_model
plot_model(model, "multi_input_and_output_model1.png", show_shapes=True)
```

```python
history = model.fit([X_train_s1,X_train_s2,X_train_s3,X_train_s4,X_train_s5,X_train_s6,X_train_s7,X_train_s8,X_train_s9,X_train_s10,X_train_s11,X_train_s12], y_train, batch_size=128, epochs=50, validation_data=([X_test_s1,X_test_s2,X_test_s3,X_test_s4,X_test_s5,X_test_s6,X_test_s7,X_test_s8,X_test_s9,X_test_s10,X_test_s11,X_test_s12] ,y_test), callbacks=[checkpoint, early_stopping])
```

```python
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len, y_vloss, marker='.', c='red', label='val_set_loss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='train_set_loss')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()
plt.show()
```
